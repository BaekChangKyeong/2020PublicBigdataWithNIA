```python
# 필요한 패키지 import
import pandas as pd
from konlpy.tag import Okt, Kkma, Komoran, Hannanum, Twitter
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.tag import untag
import xlsxwriter
```

```python
# 저장한 엑셀로부터 데이터프레임 불러오기
my_df = pd.read_excel('../data/BlueHouse.xlsx')
```

```python
# 데이터프레임 확인
print(my_df)
```

```python
# 제목 열 분리하기
title_list = my_df.제목
```

```python
# 제목 열 토큰화 진행
my_str = ''
nouns_list = []
nlpy = Okt()
for title in title_list:
    nouns = nlpy.nouns(title)
    # 행별 토큰 저장 (이중 list 객체)
    nouns_list.append(nouns)
    for j in nouns:
        # 전체 토큰 저장 (str 객체)
        my_str = my_str + ' ' + j
```

```python
# 행별 토큰 확인 (이중 list 객체)
print(nouns_list)
```

```python
# 전체 토큰 확인 (str 객체)
print(my_str)
```

```python
# 행별 토큰 엑셀로 저장
workbook = xlsxwriter.Workbook('BlueHouse_Token.xlsx')
worksheet = workbook.add_worksheet()

for i, x in enumerate(nouns_list):
    worksheet.write_row('A'+str(i+1), x)

workbook.close()
```

```python
# 전체 토큰 텍스트로 저장
fw = open('BlueHouse_Token.txt', 'w')
fw.write(my_str)
fw.close()
```
